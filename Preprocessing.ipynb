```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from sklearn.model_selection import train_test_split
```

```python
data_broadcasts = pd.read_csv("/content/drive/MyDrive/ColabNotebooks/NIR/data/ods_broadcasts_201905301157.csv")
data_broadcasts
```

```python
print(data_broadcasts.info())
```

```python
data_all_text_broadcasts = data_broadcasts.sort_values(['match_id','message_time'], ascending = True).groupby(['match_id', 'team1', 'team2', 'name', 'match_time'])['content'].apply(lambda messages: ''.join([str(message) for message in messages]).strip()).reset_index()
data_all_text_broadcasts
```

```python
data_broadcasts['content'] = data_broadcasts['content'].apply(lambda x: BeautifulSoup(str(x), 'html5lib').get_text())
```

```python
data_match_news = pd.read_csv("/content/drive/MyDrive/ColabNotebooks/NIR/data/ods_match_news.csv")
data_match_news
```

```python
print(data_match_news.info())
```

```python
data_match_news['name'] = data_match_news['name'].apply(lambda x: BeautifulSoup(str(x), 'html5lib').get_text())
data_match_news['body'] = data_match_news['body'].apply(lambda x: BeautifulSoup(str(x), 'html5lib').get_text())
```

```python
data_match_news
```

```python
data = pd.merge(data_all_text_broadcasts, data_match_news, on="match_id", how="left")
print(data.info())
```

```python
data.rename(columns = {'name_x' : 'broadcast_name', 'name_y' : 'news_name'}, inplace = True) 
data
```

```python
print(data.info())
```

```python
data['ctime'] = pd.to_datetime(data['ctime'], errors='coerce', format='%Y-%m-%d %H:%M:%S')
```

```python
import datetime
data['match_time'] = data['match_time'].apply(lambda x: datetime.datetime.fromtimestamp(x).strftime('%Y-%m-%d %H:%M:%S'))
data['match_time'] = pd.to_datetime(data['match_time'], errors='coerce', format='%Y-%m-%d %H:%M:%S')
```

```python
data.dropna(inplace=True)
```

```python
print(data.info())
```

```python
data['after_match'] = data['ctime'] - data['match_time']
```

```python
data['after_match'] = data['after_match'].apply(lambda x: 1 if x.total_seconds() > 0 else 0)
data
```

```python
data.to_csv("/content/drive/MyDrive/ColabNotebooks/NIR/data/task_class_2.csv")
```

```python
X1, X_temp = train_test_split(data, test_size=0.99, random_state=42)
```

```python
X2, X_temp = train_test_split(X_temp, test_size=0.99, random_state=42)
```

```python
X3, X_temp = train_test_split(X_temp, test_size=0.99, random_state=42)
```

```python
X4, X5 = train_test_split(X_temp, test_size=0.99, random_state=42)
```
